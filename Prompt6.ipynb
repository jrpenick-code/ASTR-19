# Sine and Cosine Functions

This notebook will define functions for sin(x) and cos(x), then tabulates and prints their values.


```python
import math

def sin_function(x):
    return math.sin(x)
```

### Sine Function

This function takes a value x and returns sin(x).


```python
def cos_function(x):
    return math.cos(x)
```

### Cosine Function

This function takes a value x and returns cos(x).


```python
num_points = 1000
x_min = 0.0
x_max = 2.0
step = (x_max - x_min) / (num_points - 1)

x_values = []
sin_values = []
cos_values = []

for i in range(num_points):
    x = x_min + i * step
    x_values.append(x)
    sin_values.append(sin_function(x))
    cos_values.append(cos_function(x))
```

```python
print("x\t\tsin(x)\t\tcos(x)")
print("------------------------------------")

for i in range(10):
    print(f"{x_values[i]:.6f}\t{sin_values[i]:.6f}\t{cos_values[i]:.6f}")
```
