```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
# Compute values
y_sin = sin_function(x)
y_cos = cos_function(x)

# Create a multipanel figure: 1 row, 2 columns
fig, axes = plt.subplots(1, 2, figsize=(8, 4))

# Left panel: sin(x)
axes[0].plot(x, y_sin)
axes[0].set_title("sin(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")

# Right panel: cos(x)
axes[1].plot(x, y_cos)
axes[1].set_title("cos(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")

# Adjust layout
plt.tight_layout()

# Save as PDF
plt.savefig("sin_cos.pdf", bbox_inches="tight")

# Show figure
plt.show()
```
